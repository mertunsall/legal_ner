## Convert json file to parquet

```python
import pandas as pd
import json

from tqdm import tqdm
```

```python
def json_to_parquet(data : list[dict],save_path) -> pd.DataFrame:

    tokenized_texts = []
    start_token_indices = []
    end_token_indices = []
    labels = []

    while len(data) > 0:
        sample = data.pop()
        tokenized_texts.append(sample["tokenized_text"])
        start_token_indices.append([ner[0] for ner in sample['ner']])
        end_token_indices.append([ner[1] for ner in sample['ner']])
        labels.append([ner[2] for ner in sample['ner']])

    data = pd.DataFrame({"tokenized_text" : tokenized_texts,
    "start_token_indices" : start_token_indices,
    "end_token_indices" : end_token_indices,
    "labels" : labels
    })

    data.to_parquet(save_path,engine='pyarrow')

    return data
    
```

```python
with open(f"anon_data/ner_data_2.json",'r') as f:
        data = json.load(f)
```

```python
json_to_parquet(data,f"anon_data/ner_data_2.parquet")
```
