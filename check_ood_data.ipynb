```python
from utils import tokenize_text, join_tokens,view, view_joined_from_dict,view_joined
import os
import json
```

```python
ood_data_path = f"ood_data/"
```

```python
def view_joined_non_incl_from_dict(item):
    tokenized_text = item['tokenized_text']
    ners = item['ner']
    ners = [[ner[0],ner[1]-1,ner[2]] for ner in ners]

    view_joined(tokenized_text,ners)
```

## names.txt

```python
with open(os.path.join(ood_data_path,"names.txt"),'r') as f:
    names = json.load(f)
```

```python
print(f"example : {names[0]}")
print(f"length of names.txt : {len(names)}")
labels_names = set()
labels_names.update([ner_label[2] for item in names for ner_label in item['ner']])
print(f"all labels in names.text : {labels_names}")
```

```python
for item in names[::5]:
    print(join_tokens(item['tokenized_text']),'\n')
```

```python
for name in names:
    view_joined_non_incl_from_dict(name)
    print('\n')
```
