```python
import pandas as pd
import re
```

```python
data_path = f"bge_data/"
```

```python
data = pd.read_parquet(data_path + f"bger-2024-2-text.parquet")
```

```python
data.head()
```

```python
pattern = "|".join(['U.________',"V.________","W.________"])

data['u'] = data["text"].str.contains("U.________",na = False,regex=False)
data['v'] = data["text"].str.contains("V.________",na = False,regex=False)
data['w'] = data["text"].str.contains("W.________",na = False,regex=False)
```

```python
texts_u = data[data['u']].sample(20,random_state = 42)["text"].tolist()
for text in texts_u:
    print(text)
```

```python
texts_v = data[data['v']].sample(20,random_state = 42)["text"].tolist()
with open("V_texte.txt",'w') as f:
    for i,text in enumerate(texts_v):
        f.write(f"Text {i+1} : \n")
        f.write(text + '\n',)
```

```python
texts_w = data[data['w']].sample(20,random_state = 42)["text"].tolist()
for text in texts_w:
    print(text)
```
